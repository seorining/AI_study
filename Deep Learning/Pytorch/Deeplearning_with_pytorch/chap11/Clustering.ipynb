```python
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
from kmeans_pytorch import kmeans, kmeans_predict
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/iris.csv')
df.info()
print('-'*60)
print(df)
```

```python
data = pd.get_dummies(df, columns=['Species'])
data['Species_Iris-setosa'] = data['Species_Iris-setosa'].apply(lambda x: 1 if x == True else 0)
data['Species_Iris-versicolor'] = data['Species_Iris-versicolor'].apply(lambda x: 1 if x == True else 0)
data
```

```python
x, y = train_test_split(data, test_size=0.2, random_state=123)
```

```python
device = 'cuda' if torch.cuda.is_available() else "cpu"
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit(data).transform(x)
y_scaled = scaler.fit(y).transform(y)
```

```python
x = torch.from_numpy(X_scaled)
y = torch.from_numpy(y_scaled)
```

```python
print(x.shape)
print(y.shape)
print(x)
```

```python
num_clusters = 3
cluster_ids_x, cluster_centers = kmeans(
    X=x, num_clusters = num_clusters, distance = 'euclidean', device=device)
```

```python
print(cluster_ids_x)
print(cluster_centers)
```

```python
cluster_ids_y = kmeans_predict(
    y, cluster_centers, 'euclidean', device=device)
```

```python
print(cluster_ids_y)
```

```python
plt.figure(figsize=(4,3), dpi=160)
plt.scatter(y[:, 0], y[:, 1], c=cluster_ids_y, cmap='viridis', marker='x')

plt.scatter(
    cluster_centers[:, 0], cluster_centers[:, 1],
    c = 'white',
    alpha = 0.6,
    edgecolors = 'black',
    linewidths=2
)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

X_train = np.load('./data/data.npy')
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(X_train)

print(gmm.means_)
print()
print(gmm.covariances_)

X, Y = np.meshgrid(np.linspace(-1, 6), np.linspace(-1, 6))
XX = np.array([X.ravel(), Y.ravel()]).T
Z = gmm.score_samples(XX)
Z = Z.reshape((50, 50))

plt.contour(X, Y, Z)
plt.scatter(X_train[:, 0], X_train[:, 1])
plt.show()
```

```python
import numpy as np
from sklearn.datasets import load_digits
from minisom import MiniSom
from pylab import plot, axis, show, pcolor, colorbar, bone

digits = load_digits()
data = digits.data
labels = digits.target
```

```python
som = MiniSom(16, 16, 64, sigma=1.0, learning_rate=0.5)
som.random_weights_init(data)
print("SOM ì´ˆê¸°í™”.")
som.train_random(data, 10000)
print("\n. SOM ì§„í–‰ ì¢…ë£Œ")

bone()
pcolor(som.distance_map().T)
colorbar()
```
