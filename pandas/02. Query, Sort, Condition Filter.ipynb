```python
df = sns.load_dataset('titanic')
df.head()
```

```python
cond1 = df['age'] >= 30
cond2 = df['who'] == 'man'

df[cond1 & cond2].sort_values(by='fare', ascending=False).head(10)
```

```python
cond1 = df['age'] >= 20
cond2 = df['age'] < 40
condition1 = cond1 & cond2
cond1 = df['pclass'] == 1
cond2 = df['pclass'] == 2
condition2 = cond1 | cond2
df.loc[condition1 & condition2, ['survived', 'pclass', 'age', 'fare']].head(10)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
condition1 = tips['day'].isin(['Fri', 'Sat'])
condition2 = tips['tip'] < 10
tips.loc[condition1 & condition2, ['total_bill', 'tip', 'smoker', 'time']].head(10)
```
